## 导入依赖

```python
import os
# - 矩阵运算
import numpy as np
# - 图像处理
import cv2 
# - 绘图可视化
from matplotlib import pyplot as plt

# 从阿凯机器人工具箱导入Gemini335类
from kyle_robot_toolbox.camera import Gemini335
```

## 相机初始化

```python
# 找到Gemini2的YAML配置文件路径
# 可以是相对路径也可以是绝对路径
yaml_path = os.path.join("config", "gemini335.yaml")
print(f"相机配置文件路径: {yaml_path}")

# 创建相机对象, 填写配置文件路径
camera = Gemini335(yaml_path=yaml_path)
```

## 图像预览

```python
# 使用read函数同时读取彩图与深度图
color_img, depth_img = camera.read()

# 生成画布
depth_canvas = camera.depth_img2canvas(depth_img, \
                                       min_distance=200, max_distance=500)
# 合并并可视化
plt.imshow(np.hstack((color_img, depth_canvas))[:, :, ::-1])
```
