## 导入依赖

```python
import os
import cv2
from matplotlib import pyplot as plt
# 阿凯机器人工具箱
# - YoloV8
from kyle_robot_toolbox.yolov8 import YoloV8Detect

%matplotlib inline
```

## 创建模型

```python
# 载入目标检测模型(使用绝对路径)
model_path = "/home/kyle/Project/YoloV8/yolov5-bottle-cap/yolov8/weights/bottle_cap_yolov8n.pt"
model = YoloV8Detect(model_path)

# 配置模型参数(可选)
# - 图像尺寸(必须是32的倍数)
model.IMAGE_SIZE= 1088
# - 置信度
model.CONFIDENCE = 0.5
# - IOU 
model.IOU = 0.6
```

## 载入图像

```python
img = cv2.imread("data/2.png")
plt.imshow(img[:, :, ::-1])
```

## 目标检测

```python
canvas, class_id_list, xyxy_list, conf_list = model.detect(img)
plt.imshow(canvas[:, :, ::-1])
```

```python
cv2.imwrite("canvas.png", canvas)
```
